```python
API_URL = "https://nlp.cst.edu.bt/nmt/api/"

CSRF_TOKEN = "cGXN7lrNRuqvjhZS6sP6WeWL77SPHBgsmc8TU2XpzlpLZZHBHO5XvfEOcLSsZYyq"


def translate_batch(sentences, count=100, delay=10, max_retries=3):

    count = 0
    
    translated_sentences = []

    for i in range(len(sentences)):
        batch = sentences[i]  

        data = {
            "csrfmiddlewaretoken": CSRF_TOKEN,  
            "src_lang": "eng_Latn", 
            "tgt_lang": "dzo_Tibt",  
            "text": batch  
        }
        
        retries = 0  

        if count == 100:
            time.sleep(50)
            count =  0
        else:
            pass


        while retries < max_retries:
            try:
                response = requests.post(API_URL, headers={"Content-Type": "application/json"}, json=data)

                if response.status_code == 200:
                    result = response.json()
                    translated_sentences.append(result)
                    count += 1
                    print(f"Translated {i} sentences...")  
                    break 

                else:
                    print(f"Error {response.status_code}: {response.text}")
                    retries += 1
                    time.sleep(5)  

            except requests.exceptions.RequestException as e:
                print(f"Request failed: {e}")
                retries += 1
                time.sleep(5)  

        with open("../Data/CST_Model_Predictions.txt", "a", encoding="utf-8") as f:
                f.write(f"{result}\n")

        time.sleep(delay)  

    return translated_sentences
```

```python
API_URL = "https://nlp.cst.edu.bt/nmt/api/"

CSRF_TOKEN = "cGXN7lrNRuqvjhZS6sP6WeWL77SPHBgsmc8TU2XpzlpLZZHBHO5XvfEOcLSsZYyq"


def translate_batch(sentences, count=100, delay=10, max_retries=3):

    count = 0
    
    translated_sentences = []

    for i in range(len(sentences)):
        batch = sentences[i]  

        data = {
            "csrfmiddlewaretoken": CSRF_TOKEN,  
            "src_lang": "dzo_Tibt", 
            "tgt_lang": "eng_Latn",  
            "text": batch  
        }
        
        retries = 0  

        if count == 100:
            time.sleep(50)
            count =  0
        else:
            pass


        while retries < max_retries:
            try:
                response = requests.post(API_URL, headers={"Content-Type": "application/json"}, json=data)

                if response.status_code == 200:
                    result = response.json()
                    translated_sentences.append(result)
                    count += 1
                    print(f"Translated {i} sentences...")  
                    break 

                else:
                    print(f"Error {response.status_code}: {response.text}")
                    retries += 1
                    time.sleep(5)  

            except requests.exceptions.RequestException as e:
                print(f"Request failed: {e}")
                retries += 1
                time.sleep(5)  

        with open("../Data/CST_Model_Predictions_Dzo_to_Eng.txt", "a", encoding="utf-8") as f:
                f.write(f"{result}\n")

        time.sleep(delay)  

    return translated_sentences
```
