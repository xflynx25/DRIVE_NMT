```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# model_checkpoint2 = "final_dzo_eng_results/checkpoint-143727"
model_checkpoint2 = "final_eng_dzo_results/checkpoint-179658"

eng_dzo_tokenizer = AutoTokenizer.from_pretrained(model_checkpoint2)
eng_dzo_model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint2)

# dzo_eng_tokenizer = AutoTokenizer.from_pretrained(model_checkpoint2)
# dzo_eng_model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint2)

if torch.cuda.is_available():
    # eng_dzo_model.to("cuda")
    eng_dzo_model.to("cuda:0")

def translate_eng_dzo(text, model = eng_dzo_model, tokenizer = eng_dzo_tokenizer, max_length=512):
    model.eval()
    inputs = tokenizer(text, return_tensors="pt", max_length=max_length, truncation=True, padding="max_length")
    
    if torch.cuda.is_available():
        inputs = inputs.to("cuda:0")
    
    with torch.no_grad():
        outputs = model.generate(inputs["input_ids"], attention_mask=inputs["attention_mask"], max_length=max_length)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return translated_text
# def translate_dzo_eng(text, model = dzo_eng_model, tokenizer = dzo_eng_tokenizer, max_length=512):
#     model.eval()
#     inputs = tokenizer(text, return_tensors="pt", max_length=max_length, truncation=True, padding="max_length")
    
#     if torch.cuda.is_available():
#         inputs = inputs.to("cuda")
    
#     with torch.no_grad():
#         outputs = model.generate(inputs["input_ids"], attention_mask=inputs["attention_mask"], max_length=max_length)
#     translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
#     return translated_text
```
