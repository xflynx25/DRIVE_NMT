```python
dzo_sentences = []
english_sentences = []

for data in ds['dev']:
    if data['glottocode'] == "dzon1239":
        dzo_sentences.append(data)
    elif data['glottocode'] == 'stan1293':
        english_sentences.append(data)
```

```python
print(len(dzo_sentences))
print(len(english_sentences))
```
