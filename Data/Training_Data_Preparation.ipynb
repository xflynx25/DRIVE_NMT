```python
import numpy as np
import pandas as pd
```

#### Preparing raw data published on DCDD Website into proper Dataframe

```python
source = []
target = []

with open('Raw/DCDD_Published_Parallel_Datasets.txt', 'r') as file:
    lines = file.readlines()

    for line in lines:
        source_target_list = str(line).split('\t')

        source_sentence = source_target_list[0].strip()

        target_sentence = source_target_list[-1].strip().split('\n')[0].strip()

        source.append(source_sentence)
        target.append(target_sentence)
```

```python
print(f'{len(target)}: {target}')
print(f'{len(source)}: {source}')
```

```python
data = pd.DataFrame({'eng':source, 'dzo':target})
```

```python
data.to_csv("Prepared/DCDD_10K_Parallel_Dataset.csv", index=False)
```
